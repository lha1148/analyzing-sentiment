```python
import pandas as pd

data = pd.read_csv("/content/drive/MyDrive/NLP/merged_hospital_reviews.csv")


def assign_sentiment_label(star):
    if "1 sao" in star or "2 sao" in star:
        return "Tệ"
    elif "3 sao" in star:
        return "Trung tính"
    elif "4 sao" in star or "5 sao" in star:
        return "Tốt"
    else:
        return "unknown"


data["sentiment"] = data["star"].apply(assign_sentiment_label)

print(data[["comment", "star", "sentiment"]])

data.to_csv(r"/content/drive/MyDrive/NLP/labeled_sentiment_data.csv", index=False)


```
